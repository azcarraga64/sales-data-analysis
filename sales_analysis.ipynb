```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales.csv", encoding="latin1")

df.head()
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

category_sales
```

```python
category_sales.plot(kind="bar", title="Sales by Category")

plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

region_sales
```

```python
region_sales.plot(kind="bar", title="Sales by Region")

plt.show()
```

```python
top_products = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)

top_products
```

```python
top_products.plot(kind="bar", title="Top 10 Products")

plt.show()
```

## Insights

- The Technology category generates the highest revenue.
- The West region shows the strongest sales performance.
- A small group of products contributes to a large portion of total sales.
